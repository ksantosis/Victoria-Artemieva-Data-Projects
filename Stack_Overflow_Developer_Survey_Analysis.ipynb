Statistical analysis of respondents' experience

```python
mean   = df['WorkExp'].mean()
median = df['WorkExp'].median()
mode   = df['WorkExp'].mode()[0] if not df['WorkExp'].mode().empty else None

print(f"Mean: {mean:.2f} years")
print(f"Median: {median:.2f} years")
print(f"Mode: {mode:.2f} years")
```

Remote work analysis

```python
# search for column containing 'remote'
for col in df.columns:
    if 'remote' in col.lower():
        print(col)

print(df['RemoteWork'].unique())  # see available options

remote_count = df['RemoteWork'].value_counts().get('Remote', 0)
print("\nNumber of respondents working remotely:", remote_count)
```

Python popularity

```python
# search for language columns
for col in df.columns:
    if 'language' in col.lower():
        print(col)

# calculate Python usage percentage
python_users = df['LanguageHaveWorkedWith'].str.contains('Python', na=False).sum()
python_popularity = (python_users / len(df)) * 100
print(f"Respondents who program in Python: {python_popularity:.1f}%")
```

Learning paths

```python
# search for learning related columns
for col in df.columns:
    if 'learn' in col.lower() or 'course' in col.lower():
        print(col)

online_courses_count = df['LearnCode'].str.contains('Online Courses', na=False, case=False).sum()
print(f"Number of respondents who learned through online courses: {online_courses_count}")
```

Compensation by country for Python developers

```python
python_devs = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)

compensation_by_country = (df[python_devs]
    .dropna(subset=['ConvertedCompYearly'])
    .groupby('Country')['ConvertedCompYearly']
    .agg(Mean_Compensation='mean', Median_Compensation='median')
    .round(2))

display(compensation_by_country.head(20))
```

Education of top 5 highest paid

```python
top_5 = df.sort_values(by='ConvertedCompYearly', ascending=False).head(5)[['ConvertedCompYearly', 'EdLevel']]
print("Education of the top 5 highest-paid respondents:")
display(top_5)
```

Python popularity by age group

```python
python_by_age = df[python_devs].groupby('Age').size()
total_by_age = df.groupby('Age').size()

result = (python_by_age / total_by_age * 100).round(1)
result = result.to_frame(name='Python_usage_%')
display(result)
```

Industries of high-paid remote workers

```python
percentile_75 = df['ConvertedCompYearly'].quantile(0.75)

top_earners_remote = df[
    (df['ConvertedCompYearly'] >= percentile_75) &
    (df['RemoteWork'] == 'Remote')
]

industry_count = top_earners_remote['Industry'].value_counts().head(10)
print("\nIndustries among high-paid remote workers:")
display(industry_count)
```
